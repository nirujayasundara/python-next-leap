# Python String Methods

Strings are immutable sequences of characters.
Below are commonly used string methods with examples and outputs.


```

s = "hello world"

print("capitalize:", "hello".capitalize())
print("casefold:", "HELLO".casefold())
print("center:", "hi".center(10,'*'))
print("count:", s.count("o"))
print("encode:", "hello".encode())
print("endswith:", s.endswith("world"))
print("expandtabs:", "hi\tthere".expandtabs(4))
print("find:", s.find("world"))
print("format:", "Hello {}".format("Aditya"))
print("format_map:", "Hi {name}".format_map({"name":"Aditya"}))
print("index:", s.index("world"))
print("isalnum:", "abc123".isalnum())
print("isalpha:", "abc".isalpha())
print("isascii:", "abc".isascii())
print("isdecimal:", "123".isdecimal())
print("isdigit:", "123".isdigit())
print("isidentifier:", "var_1".isidentifier())
print("islower:", "hello".islower())
print("isnumeric:", "123".isnumeric())
print("isprintable:", "hello".isprintable())
print("isspace:", "   ".isspace())
print("istitle:", "Hello World".istitle())
print("isupper:", "HELLO".isupper())
print("join:", "-".join(["a","b"]))
print("ljust:", "hi".ljust(6,'*'))
print("lower:", "HELLO".lower())
print("lstrip:", "  hi".lstrip())
print("partition:", "hello world".partition(" "))
print("removeprefix:", "unhappy".removeprefix("un"))
print("removesuffix:", "file.txt".removesuffix(".txt"))
print("replace:", "hello".replace("l","x"))
print("rfind:", s.rfind("o"))
print("rindex:", s.rindex("o"))
print("rjust:", "hi".rjust(6,'*'))
print("rpartition:", "hello world".rpartition(" "))
print("rsplit:", "a,b,c".rsplit(",",1))
print("rstrip:", "hi   ".rstrip())
print("split:", "a,b,c".split(","))
print("splitlines:", "hi\nhello".splitlines())
print("startswith:", s.startswith("hello"))
print("strip:", "  hi  ".strip())
print("swapcase:", "Hello".swapcase())
print("title:", "hello world".title())
print("translate:", "abc".translate(str.maketrans("a","x")))
print("upper:", "hello".upper())
print("zfill:", "5".zfill(3))
```


# Tuple Methods

Tuples are immutable ordered collections.
They only have two built-in methods.


```

t = (1,2,3,2)
print("count:", t.count(2))
print("index:", t.index(3))
```


# List Methods

Lists are mutable ordered collections.
They support many built-in methods.


```

lst = [1,2,3]
lst.append(4)
print("append:", lst)
lst.insert(1,100)
print("insert:", lst)
lst.extend([5,6])
print("extend:", lst)
print("index:", lst.index(100))
print("count:", lst.count(2))
lst.remove(2)
print("remove:", lst)
print("pop:", lst.pop())
lst.reverse()
print("reverse:", lst)
lst.sort()
print("sort:", lst)
```


# Dictionary Methods

Dictionaries store key-value pairs.
Below are common dictionary methods.


```

d = {"a":1,"b":2}
print("get:", d.get("a"))
print("keys:", d.keys())
print("values:", d.values())
print("items:", d.items())
d.update({"c":3})
print("update:", d)
print("pop:", d.pop("a"))
print("setdefault:", d.setdefault("d",4))
print("fromkeys:", dict.fromkeys(["x","y"],0))
```


# Set Methods

Sets are unordered collections of unique elements.
They support mathematical set operations.


```

s = {1,2,3}
s.add(4)
print("add:", s)
print("union:", s.union({5,6}))
print("intersection:", s.intersection({2,3,5}))
print("difference:", s.difference({2}))
print("issubset:", {1,2}.issubset(s))
print("issuperset:", s.issuperset({1}))
```
